```python
import pandas as pd
```

```python
df = pd.read_csv('pandas/data/bestsellers.csv')
```

```python
df
```

```python
df.plot?
```

```python
df.plot(kind='scatter', x='Reviews', y='Price');
```

```python
df.plot(kind='scatter',
        x='Reviews',
        y='Price',
        color='orange',
        title='Reviews vs. Price',
        figsize=(12, 6))
```

Bar Plot

```python
df_year = df.groupby('Year', as_index=False).mean(numeric_only=True).reset_index()
df_year
```

```python
df_year.plot(
    kind="bar",
    x='Year',
    y='Reviews',
)
```

```python
df_year.plot(kind='bar',
             x='Year',
             y='Reviews',
             color='lightgreen',
             title='Reviews per Year',
             figsize=(12, 6))
```

Also is possible with columns without numeric values

```python
top_authors = df['Author'].value_counts().head(10)
top_authors
```

```python
top_authors.plot(kind='bar',
                 color='lightblue',
                 title='Top book quantity per Author',
                 figsize=(12, 6))
```

Line Plot

```python
df_year['User Rating']
```

```python
df_year['User Rating'].plot(kind='line')
```

```python
df_year['User Rating'].plot(color='purple',
                            linewidth=4,
                            title='Average User Rating per Year',
                            figsize=(12, 6))
```
